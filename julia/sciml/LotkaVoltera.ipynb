```julia
using Plots
using DifferentialEquations
using ParameterizedFunctions
```

```julia
lv! = @ode_def LotkaVoltera begin
    dx = a*x - b*x*y
    dy = -c*y + d*x*y
end a b c d
```

```julia
u0 = [1.0, 1.0]
p = (1.5, 1.0, 3.0, 1.0)
tspan = (1.0, 10.0)
prob = ODEProblem(lv!, u0, tspan, p)
```

```julia
sol = solve(prob)
```

```julia
plot(sol)
```
