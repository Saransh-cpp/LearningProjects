```julia
using Plots
using DifferentialEquations
using ParameterizedFunctions
```

```julia
van! = @ode_def VanDerPol begin
    dy = μ * ((1 - x^2) * y - x)
    dx = y
end μ
```

```julia
prob = ODEProblem(van!, [0.0, 2.0], (0.0, 6.3), 1e6)
```

```julia
sol = solve(prob)
```

```julia
plot(sol)
```
